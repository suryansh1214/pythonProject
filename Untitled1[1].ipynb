```python
#creating a copy of original df for further use
from sklearn.preprocessing import LabelEncoder

# Fit the LabelEncoder instance to the 'localGoals' column
label_encoder = LabelEncoder()
label_encoder.fit(df['localGoals'])

# Fit the LabelEncoder instance to the 'visitorGoals' column
le_transmission = LabelEncoder()
le_transmission.fit(df['visitorGoals'])

# Now you can inverse transform the columns
namedDf = df.copy()
namedDf['localGoals'] = label_encoder.inverse_transform(label_encoder.transform(df['localGoals']))
namedDf['visitorGoals'] = le_transmission.inverse_transform(le_transmission.transform(df['visitorGoals']))

namedDf.head(1)
```

```python
# Histogram
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(5,3))
sns.histplot(namedDf['timestamp'], kde = True)
plt.title('Distribution of Football Timestamp')
plt.xlabel('timestamp')
plt.ylabel('Count')
plt.show()
```

```python
x = df.iloc[:,:-1] #Independent Variable / Features
y = df.iloc[:,-1] #Dependent Variable / Target value

# Train and Test Data Split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25, random_state = 100)
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler

# Scaling of the Data
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
# Model Selection and Building
from sklearn.linear_model import LinearRegression
regression = LinearRegression(n_jobs= -1)

# model Training
regression.fit(x_train,y_train)
```

```python
# Multi Linear Regression Model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

regression = LinearRegression(n_jobs= -1)
regression.fit(x_train,y_train)
y_pred = regression.predict(x_test)

score1 = r2_score(y_test,y_pred)
print("R2 Score: ", score1)
```

```python
# Decision Tree Regression Model
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor()
dt.fit(x_train,y_train)
y_pred2 = dt.predict(x_test)

score2 = r2_score(y_test, y_pred2)
print("R2 Score: ", score2)
```

```python
# Random Forest Regression Model
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()
rf.fit(x_train,y_train)
y_pred3 = rf.predict(x_test)

score3 = r2_score(y_test, y_pred3)
print("R2 Score: ", score3)
```

```python
# Comparison of Different Models Predictionâ€‹

a = pd.DataFrame({"Algorithm":['Linear Regressor','Decision_Tree','Random-Forest'],
                  "Accuracy": [score1,score2,score3],
                  "prediction-1":[y_pred[1],y_pred2[1],y_pred3[1]]},)
a
```
